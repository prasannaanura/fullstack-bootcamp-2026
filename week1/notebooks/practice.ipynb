```python
numbers = [1,2,3,4,5,6]
sqaures = []
for num in numbers:
    sqaures.append(num ** 2)
print(sqaures)
```

```python
squares = [num ** 2 for num in  numbers]
print(sqaures)
```

```python
def add_num(a,b):
 return a ** b

result = add_num(5,2)
print(result)
```

```python
sqaures = lambda a,b: a**b
result = sqaures(5,2)
print(result)
```

```python
even_num = [num for num in range(100) if num % 2 == 0]
print(even_num)
print(len(even_num))
```

```python
sqaures_dic = {num: num**2 for num in range(1,10)}
print(sqaures_dic)
```

```python
def double(x):
    return x * 2

numbers = [1, 2, 3]

mapped = map(double,numbers)
result = list(mapped)
print(result)
```
