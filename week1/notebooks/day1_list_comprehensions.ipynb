## CHALLENGE #1: First List Comprehension

```python
#traditional way with loop
numbers = [1,2,3,4,5]
sqaures = []
for num in numbers:
    sqaures.append(num**2)
print("using loops",sqaures)
print("")
#list comprehension
numbers = [1,2,3,4,5]
squares_com = [num ** 2 for num in numbers]
print(f"list comprehension{squares_com}")
```
