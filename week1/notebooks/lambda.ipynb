## Lambda function vs traditional function

```python
def squares(x):
    return x ** 2
result = squares(9)
print(result)
```

```python
squares = lambda x: x ** 2
result = squares(5)
print(result)

```

```python
def add_ten(a):
    return a + 10
print("Traditional",add_ten(5))

add_ten = lambda a: a + 10
print("Lambda way",add_ten(5))
```

```python
add =lambda a,b: a*b
print(add(2,5))
```
